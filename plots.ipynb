```python
import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
imp_df = pd.read_csv('data_imp.csv', usecols=[1,2,3,4,5,6,7,17,18])
imp_df = imp_df.dropna()
gen_df = pd.read_csv('generated_data.csv')
```

Compare distribution of generated and original data

```python
fig = plt.figure(figsize=(20,20))

for i, col in enumerate(imp_df.columns):
    ax1 = fig.add_subplot(5,5,i+1)
    ax1.hist(imp_df[col], label='real')
    ax1.hist(gen_df[col], alpha=0.5, label='gen')
    ax1.set_title(col)
    plt.legend()
    
plt.show()

```

PCA plots for generated and original data (first 2 principal components)

```python
pca = PCA(n_components=2)
pca2 = PCA(n_components=2)
#pca.fit(imp_df)
imp_pca = pca.fit_transform(imp_df)
gen_pca = pca.transform(gen_df)
plt.scatter(imp_pca[:, 0], imp_pca[:, 1], label='real')
plt.scatter(gen_pca[:, 0], gen_pca[:, 1], label='gen')
plt.xlabel('first principal component')
plt.ylabel('second principal component')
plt.legend()
```
